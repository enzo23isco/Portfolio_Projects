## Body Recomposition Phase - (September - December)

```python
# Get the user's data using input function

goal_input = input("Please enter goal (surplus, deficit, recom) : ")
weight_input = input("Enter weight (kg): ")
height_input = input("Enter height (cm): ")
age_input = input("Enter your age: ")
 

# Convert input values to numeric values (float, int)

weight = float(weight_input)
height = float(height_input)
age = int(age_input)

# Convert Goal input into all lower case

goal = goal_input.lower()

# Calculate Basal Metabolism - Mifflin St Jeor Formula Method

basal_metabolism = (10 * weight + 6.25 * height - 5 * age) + 5

print("Your basal metabolism value is: ", basal_metabolism)
```

```python
## Here I'll explain to the user 

activity_input = input("Enter activity factor (sedentary, light, moderate, intense): ")
activity = activity_input.lower()

# if statements to create values based on activity input

if(activity == "sedentary"):
    TEE = basal_metabolism * 1.2
elif(activity == "light"):
    TEE = basal_metabolism * 1.375
elif(activity == "moderate"):
    TEE = basal_metabolism * 1.55
elif(activity == "intense"):
    TEE = basal_metabolism * 1.725
else:
    print("Enter a valid input (see instructions)")
    
print("The Total Energetic Expenditure (TEE) in calories per day is:", TEE)     
```

```python
# Final calculation for the Goal

if(goal == "surplus"):
    y = 150
    caloric_surplus = TEE + y
    print("Total calories needed for caloric surplus are:", caloric_surplus)
    
    # Protein Macronutrient Intake
    protein_grams = 1.65 * weight
    total_caloric_protein = protein_grams * 4
    print("Protein grams needed:", protein_grams)
    ### print("Total protein calories needed (after conversion):", total_caloric_protein)
    
    # Fat Macronutrient Intake
    fat = 0.25 * TEE
    fat_grams = fat/9
    print("Fat grams needed:", fat_grams)
    ### print("Fat grams calories needed (after conversion):", fat_grams)
    
    # Total carbs intake
    total_carbs = caloric_surplus - protein_grams - fat
    carbs_grams = total_carbs/4
    print("Carbs grams needed:", carbs_grams)
    
elif(goal == "deficit"):
    y = 450
    caloric_deficit = TEE - y
    print("Total calories needed for caloric deficit are:", caloric_deficit)
    
    # Protein Macronutrient Intake
    protein_grams = 1.65 * weight
    total_caloric_protein = protein_grams * 4
    print("Protein grams needed:", protein_grams)
    ### print("Total protein calories needed (after conversion):", total_caloric_protein)
    
    # Fat Macronutrient Intake
    fat = 0.25 * TEE
    fat_grams = fat/9
    print("Fat grams needed:", fat_grams)
    ### print("Fat grams calories needed (after conversion):", fat_grams)
    
    # Total carbs intake
    total_carbs = caloric_deficit - protein_grams - fat
    carbs_grams = total_carbs/4
    print("Carbs grams needed:", carbs_grams)
    
elif(goal == "recom"):
    y = TEE * 0.10
    caloric_deficit = TEE - y 
    print("Total calories needed for body recomposition are:", caloric_deficit)
    
     # Protein Macronutrient Intake
    protein_grams = 2 * weight
    total_caloric_protein = protein_grams * 4
    print("Protein grams needed:", protein_grams)
    print("Total protein calories needed (after conversion):", total_caloric_protein)
    
    # Fat Macronutrient Intake
    fat = 0.25 * caloric_deficit
    fat_grams = fat/9
    print("Fat grams needed:", fat_grams)
    print("Fat grams calories needed (after conversion):", fat)
    
    # Total carbs intake
    total_carbs = caloric_deficit - total_caloric_protein - fat
    carbs_grams = total_carbs/4
    print("Carbs grams needed:", carbs_grams)
    print("Total Carb grams calories needed (after conversion:", total_carbs)
    
    # 
    
    #print(
    
else:
    print("Enter a valid input")
    
```
